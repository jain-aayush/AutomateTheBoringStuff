```python
"""This is a compilation of data analysis and visualization done on the FIFA17 data set to arrive at a possible 
playing 11 in the 4-3-3 formation. 
we select a Goalkeeper
4 defenders - 1 left defender, 1 right defender and 2 centre backs
3 mid-fielders - 1 left mid-fielder, 1 right mid-fielder and 1 central mid-fielder
3 strikers - 1 left winger, 1 right winger and 1 centre-forward

The positions are selected as per the national position given in the data set.
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
% matplotlib inline
df = pd.read_csv("FullData.csv")
```

```python
#assigning weights for various analysis
a = 0.5
b = 1
c = 1.5
d = 2
```

```python
#GoalKeeping statistics
df['GoalKeeping_Skills'] = (b*df.Balance + c* df.Agility + c*df.Jumping + d*df.GK_Positioning + d*df.GK_Diving + b*df.GK_Kicking + c*df.GK_Handling + d*df.GK_Reflexes)/(2*b + 3*c+ 3*d)
```

```python
plt.figure(figsize = (15,6))

sd1 = df.sort_values('GoalKeeping_Skills',ascending = False)[:5]
x1= np.array(list(sd1['Name']))
y1 = np.array(list(sd1['GoalKeeping_Skills']))
sns.barplot(x1, y1, palette = 'colorblind')
plt.ylabel("GoalKeeping Skills Score")
```

```python
#choosing defender
df['Defending_Skills'] = (d*df.Marking + b*df.Ball_Control + d*df.Sliding_Tackle + d*df.Standing_Tackle + a*df.Aggression + d*df.Interceptions
                         + c*df.Vision + c*df.Composure + b*df.Crossing + a*df.Short_Pass + a*df.Long_Pass
                         + b*df.Strength + c*df.Balance + a*df.Jumping)/(4*a + 3*b + 3*c+ 4*d)
```

```python
plt.figure(figsize = (15,6))
sd1 = df.sort_values('Defending_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'LB'][:5]
x2 = np.array(list(sd2['Name']))
y2 = np.array(list(sd2['Defending_Skills']))
sns.barplot(x2, y2, palette = sns.color_palette('Blues_d'))
plt.ylabel("Left Back Defender Score")
```

```python
plt.figure(figsize = (15,6))
sd1 = df.sort_values('Defending_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'RB'][:5]
x2 = np.array(list(sd2['Name']))
y2 = np.array(list(sd2['Defending_Skills']))
sns.barplot(x2, y2, palette = sns.color_palette('Blues_d'))
plt.ylabel("Right Back Defender Score")
```

```python
plt.figure(figsize = (15,6))
sd1 = df.sort_values('Defending_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'CB'][:5]
x2 = np.array(list(sd2['Name']))
y2 = np.array(list(sd2['Defending_Skills']))
sns.barplot(x2, y2, palette = sns.color_palette('Blues_d'))
plt.ylabel("Centre Back Defender Score")
```

```python
#calculating midfeild statistics

df['Midfielding_Skills'] = (d*df.Ball_Control + c*df.Dribbling + a*df.Marking + a*df.Standing_Tackle + a*df.Sliding_Tackle
                            + b*df.Interceptions + b*df.Composure + c*df.Short_Pass + c*df.Long_Pass 
                           + d*df.Stamina + d*df.Strength + a*df.Jumping + b*df.Shot_Power)/(4*a + 3*b+ 3*c + 3*d)
```

```python
plt.figure(figsize = (15,6))

sd1 = df.sort_values('Midfielding_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'LM'][:5]
x3 = np.array(list(sd2['Name']))
y3 = np.array(list(sd2['Midfielding_Skills']))
sns.barplot(x3, y3, palette = sns.color_palette('Blues_d'))
plt.ylabel("Left Midfielder Score")
```

```python
plt.figure(figsize = (15,6))

sd1 = df.sort_values('Midfielding_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'CAM'][:5]
x3 = np.array(list(sd2['Name']))
y3 = np.array(list(sd2['Midfielding_Skills']))
sns.barplot(x3, y3, palette = sns.color_palette('Blues_d'))
plt.ylabel("Centre Midfielder Score")
```

```python
plt.figure(figsize = (15,6))

sd1 = df.sort_values('Midfielding_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'RM'][:5]
x3 = np.array(list(sd2['Name']))
y3 = np.array(list(sd2['Midfielding_Skills']))
sns.barplot(x3, y3, palette = sns.color_palette('Blues_d'))
plt.ylabel("Right Midfielder Score")
```

```python
#calculating striking statistics
df['GoalScoring_Skills'] = (c*df.Ball_Control + c*df.Dribbling + d*df.Attacking_Position + a*df.Short_Pass
                            + b*df.Acceleration + d*df.Speed + a*df.Stamina + b*df.Jumping + c*df.Heading
                          + d*df.Shot_Power + d*df.Finishing + b*df.Long_Shots + b*df.Curve + b*df.Freekick_Accuracy
                           + b*df.Penalties)/(2*a + 6*b + 3*c + 4*d)
```

```python
plt.figure(figsize = (15,6))

sd1 = df.sort_values('GoalScoring_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'LW'][:5]
x4 = np.array(list(sd2['Name']))
y4 = np.array(list(sd2['GoalScoring_Skills']))
sns.barplot(x4, y4, palette = sns.color_palette('Blues_d'))
plt.ylabel("Left Winger Score")
```

```python
plt.figure(figsize = (15,6))

sd1 = df.sort_values('GoalScoring_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'ST'][:5]
x4 = np.array(list(sd2['Name']))
y4 = np.array(list(sd2['GoalScoring_Skills']))
sns.barplot(x4, y4, palette = sns.color_palette('Blues_d'))
plt.ylabel("Centre-Forward Score")
```

```python
plt.figure(figsize = (15,6))

sd1 = df.sort_values('GoalScoring_Skills',ascending = False)
sd2 = sd1.loc[sd1['National_Position'] == 'RW'][:5]
x4 = np.array(list(sd2['Name']))
y4 = np.array(list(sd2['GoalScoring_Skills']))
sns.barplot(x4, y4, palette = sns.color_palette('Blues_d'))
plt.ylabel("Right Winger Score")
```
